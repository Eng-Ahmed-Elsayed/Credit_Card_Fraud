## Import libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load the data

```python
df = pd.read_csv('creditcard.csv') 
df.head()
```

## Preprocess the data

```python
df.info()
# No Missing Values
```

```python
df.describe()
```

```python
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
y = df['Class'].values
X = df.drop(['Class'], axis = 1)
X = preprocessing.StandardScaler().fit(X).transform(X)
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.2, random_state=4)
```

## Logistic Regression

```python
from sklearn.metrics import f1_score
from sklearn.metrics import jaccard_score
from sklearn.metrics import classification_report, confusion_matrix
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import log_loss
lr = LogisticRegression(C=0.001, solver='newton-cg').fit(X_train, y_train)
lr.fit(X_train, y_train)
yhat_lr = lr.predict(X_test)
yhat_prob = lr.predict_proba(X_test) 
print(f"log_loss: {round(log_loss(y_test, yhat_prob), 4)}")
print(f"f1_score = {round(f1_score(y_test, yhat_lr, average='weighted'), 4)}")
print(f"jaccard_score =  {round(jaccard_score(y_test, yhat_lr, average='weighted', pos_label=1), 4)}")
```

## Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
maxDepths = 10
acc_score = np.zeros((maxDepths))
f1Tree = np.zeros((maxDepths))
jacTree = np.zeros((maxDepths))
for maxDepth in range(1, maxDepths+1):
    creditTree = DecisionTreeClassifier(criterion="entropy", max_depth=maxDepth)
    creditTree.fit(X_train, y_train)
    treePredict = creditTree.predict(X_test)
    acc_score[maxDepth-1] = metrics.accuracy_score(y_test, treePredict)
    f1Tree[maxDepth-1] = f1_score(y_test, treePredict, average='weighted')
    jacTree[maxDepth-1] = jaccard_score(y_test, treePredict, average='weighted', pos_label=1)

print( f"maxDepth={acc_score.argmax()+1 } has the best accuracy = {round(acc_score.max(), 3)}") 
print( f"maxDepth={f1Tree.argmax()+1 } has the best f1_score = {round(f1Tree.max(), 3)}") 
print( f"maxDepth={jacTree.argmax()+1 } has the best jaccard_score = {round(jacTree.max(), 3)}") 
loanTree = DecisionTreeClassifier(criterion="entropy", max_depth=f1Tree.argmax()+1).fit(X_train, y_train)
```
